```python
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio
from google.colab import files
```

```python
mytext = input().split()
mytext = ', ,'.join(mytext)
audio = gTTS(text = mytext,
                    lang = 'en',
                    tld = 'co.in',
                    slow = False
                    )

audio.save("/content/gtts.wav")
Audio("/content/gtts.wav")
```
